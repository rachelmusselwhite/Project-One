```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
wage_requests = requests.get("https://en.wikipedia.org/wiki/List_of_US_states_by_minimum_wage#Notes")
wage = wage_requests.text
print(wage)
```

```python

soup = BeautifulSoup(wage)
type(soup)
```

## 
First we will grab the most recent table from the wikipage

```python
recent_body = soup.select(".wikitable")[0]
print(recent_body.text.strip()[:100])
```

## Now we will select the row of column headers


```python
rows = recent_body.select("tr")

first_row = rows[0]
first_row
```

## Now we extract names

```python
header_names = []

for name in first_row.select("th"):
    title = name.text.strip()
    header_names.append(title)
header_names

#we will need to remove footnotes from "2024"
```

## Now we select the elements from the table

```python
len(recent_body.select("tr")[1:])
```

```python
recent_elements = recent_body.select("tr")[1:]
```

## Each element now goes into a python list

```python
recent_elements[0].select("td")
```

```python
recent_body.select("tr")[1].select("td")
```

```python
states_list= []
for item in recent_body.select("tr")[1:]:
    state_list = []
    for td in item.select("td"):
        state_list.append(td.text.strip())
    print(state_list)
    states_list.append(state_list)
```

```python
states_list
```

## And now a dataframe is created

```python
df = pd.DataFrame(states_list, columns=["State", "2022", "2023", "2024"])
df
```

```python
df['2024'] = df['2024'].str.replace('(Federal, no state minimum)', '').str.replace('(Employers subject to the Fair Labor Standards Act must pay the $7.25 federal minimum wage)', '')
```

```python
df['2023'] = df['2023'].str.replace('(Federal, no state minimum)', '').str.replace('(Employers subject to the Fair Labor Standards Act must pay the $7.25 federal minimum wage)', '')
```

```python
df['2022'] = df['2023'].str.replace('(Federal, no state minimum)', '').str.replace('(Employers subject to the Fair Labor Standards Act must pay the $7.25 federal minimum wage)', '')
```
